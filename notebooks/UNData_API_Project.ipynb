```python
import requests
import json
import pandas as pd
```

```python
#reference for the future: endpoint url is a good opportunity for f strings (easily change between country name and indicator name)
endpoint_GDP = 'http://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD'
params = {
    "format":"json",
    "Per_page":16758
}
```

```python
response_GDP= requests.get(endpoint_GDP,params)
response_GDP.json()[0]
```

```python
gdp_df_raw = pd.json_normalize(response_GDP.json()[1])
gdp_df_raw.tail()
```

```python
endpoint_Life_Expectancy = 'http://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN'
params = {
    "format":"json",
    "Per_page":16758
}
```

```python
response_Life_Expectancy = requests.get(endpoint_Life_Expectancy,params)
response_Life_Expectancy.json()[0]
```

```python
LE_df_raw = pd.json_normalize(response_Life_Expectancy.json()[1])
LE_df_raw.tail()
```

```python
gdp_df = (gdp_df_raw
          .rename(columns = {"value":"GDP_Per_Capita"})
          .drop(columns = ["unit","obs_status","decimal","indicator.id","indicator.value","country.id"])
          .copy()
         )
gdp_df.tail()
```

```python
LE_df = (LE_df_raw
         .rename(columns = {"value":"Life_Expectancy_Years"})
         .drop(columns = ["unit","obs_status","decimal","indicator.id","indicator.value","country.id"])
         .copy()
        )
LE_df.tail()
```

```python
gdp_le_raw = gdp_df.merge(LE_df,how='inner',on=["country.value","date"])
gdp_le_raw.tail()
```

```python
gdp_le = (gdp_le_raw
          .drop(columns = ["countryiso3code_y"])
          .rename(columns = {"countryiso3code_x":"Country_Code","date":"Year","country.value":"Country"})
          .copy()
         )
gdp_le.tail()
```

```python
endpoint_country = 'http://api.worldbank.org/v2/country'
params = {
    "format":"json",
    "Per_page":297
}

response_country = requests.get(endpoint_country, params)
response_country.json()[0]
```

```python
country_df_raw = pd.json_normalize(response_country.json()[1])
country_df_raw.tail()
```

```python
country_df = (country_df_raw
             .rename(columns = {'name':'Country'})
             .copy()
             )
country_df.tail()
```

```python
gdp_le_country = gdp_le.merge(country_df,on='Country',how='outer')
print(gdp_le.shape)
print(gdp_le_country.shape)
gdp_le_country.tail()
```

```python
gdp_le_countries_only = (gdp_le_country
                         .drop(index=gdp_le_country[gdp_le_country['region.value']=='Aggregates'].index)
                         .copy()
                        )
print(gdp_le_countries_only.shape)
gdp_le_countries_only.tail()
```

```python
endpoint_GDP = 'http://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD'
page = 1
params = {
    "format":"json",
    "Page":page
}
    
new_response_gdp = requests.get(endpoint_GDP,params)
new_gdp_df = pd.json_normalize(new_response_gdp.json()[1])

while True:
    page+=1
    params = {
        "format":"json",
        "Page":page
    }
    new_response_gdp = requests.get(endpoint_GDP,params)
    if pd.json_normalize(new_response_gdp.json()[1]).empty:
        break
    else:
        new_gdp_df = pd.concat([new_gdp_df,pd.json_normalize(new_response_gdp.json()[1])])

print(new_gdp_df.shape)
new_gdp_df.tail()
```
