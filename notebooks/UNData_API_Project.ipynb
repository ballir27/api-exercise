```python
import requests
import json
import pandas as pd
```

```python
endpoint_GDP_2017 = 'http://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD'
params = {
    "format":"json",
    "Per_page":16758
}
```

```python
response_GDP= requests.get(endpoint_GDP_2017,params)
response_GDP.json()[0]
```

```python
gdp_df_raw = pd.json_normalize(response_GDP.json()[1])
gdp_df_raw.tail()
```

```python
endpoint_Life_Expectancy = 'http://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN'
params = {
    "format":"json",
    "Per_page":16758
}
```

```python
response_Life_Expectancy = requests.get(endpoint_Life_Expectancy,params)
response_Life_Expectancy.json()[0]
```

```python
LE_df_raw = pd.json_normalize(response_Life_Expectancy.json()[1])
LE_df_raw.tail()
```

```python
gdp_df = (gdp_df_raw
          .rename(columns = {"value":"GDP_Per_Capita"})
          .drop(columns = ["unit","obs_status","decimal","indicator.id","indicator.value","country.id"])
          .copy()
         )
gdp_df.tail()
```

```python
LE_df = (LE_df_raw
         .rename(columns = {"value":"Life_Expectancy_Years"})
         .drop(columns = ["unit","obs_status","decimal","indicator.id","indicator.value","country.id"])
         .copy()
        )
LE_df.tail()
```

```python
gdp_le_raw = gdp_df.merge(LE_df,how='inner',on=["country.value","date"])
gdp_le_raw.tail()
```

```python
gdp_le = (gdp_le_raw
          .drop(columns = ["countryiso3code_y"])
          .rename(columns = {"countryiso3code_x":"Country_Code","date":"Year","country.value":"Country"})
          .copy()
         )
gdp_le.tail()
```
